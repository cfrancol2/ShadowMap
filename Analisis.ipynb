```python
columnas_str = ['id_mensaje', 'huella_contenido', 'id_hilo', 'usuario', 'fecha_hora', 'titulo_limpio', 'cuerpo_limpio',
                    'texto_citado_limpio', 'url_original']
for col in columnas_str:
    if col in datos.columns:
        datos[col] = datos[col].fillna('').astype("string")
```

```python
# En pandas, las columnas de texto suelen mostrarse con dtype "object"
# aunque cada valor sea una cadena. Por eso `astype(str)` no cambia la etiqueta
# de dtype a "str" en el resumen de `info()`.
# Si quieres usar el dtype de cadena de pandas, convierte a "string":
datos[columnas_str] = datos[columnas_str].astype("string")
datos[columnas_str].dtypes
```
